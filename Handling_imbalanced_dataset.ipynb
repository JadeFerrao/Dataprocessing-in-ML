```python
# separating the legit and fraudulent trinsactions
legit = credit_card_data[credit_card_data.Class == 0]
fraud = credit_card_data[credit_card_data.Class == 1]
```

```python
print (legit.shape)
print (fraud.shape)
```

Under-sampling  

Build a sample dataset containing similar distribution of Legit & Fraudulent Transactionds

```python
legit_sample = legit.sample (n=2) # since 2 imbalanced data
```

```python
print (legit_sample.shape)
```

Concatenate the Two DataFrames

```python
new_dataset = pd.concat([legit_sample, fraud], axis = 0) #concatenated row wise
```
